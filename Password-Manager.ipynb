```python
import hashlib #Importing hashlib library for securly password hashing password using SHA-256 encrytionp
import os #Importing OS library for file handling 

#********************Hashing of the password using SHA-256 algorithm***********************
def hash_password(password):
    
    sha256 = hashlib.sha256() # To create a new SHA-256 hash object
    sha256.update(password.encode()) #Updating SHA-256 object in bytes 
    hashed_password = sha256.hexdigest() #Hexadecimal representation of the password
    return hashed_password

 #********************Encryption Function**************************************************
def encrypt_password(password, encryption_key):
    #The zip function accepts password and encryption_key and return overall tuple
    #XOR operation takes place between the tuple that is returned by the zip function and then converted to the byte
    encrypted_password = bytes([p ^ k for p, k in zip(password, encryption_key)]) 
    return encrypted_password

#********************Decrytion Function***************************************************
def decrypt_password(encrypted_password, encryption_key):
    #The zip function accepts encrypted_password and encryption_key and return overall tuple
    #XOR operation takes place between the tuple that is returned by the zip function and then converted to the byte
    decrypted_password = bytes([e ^ k for e, k in zip(encrypted_password, encryption_key)])
    return decrypted_password

#*******************Password additon Function**********************************************
def add_password(username, password, encryption_key):
    try:
        encrypted_password = encrypt_password(password.encode(), encryption_key)
        with open("passwords.txt", "ab") as file: #Use of context manager for reading the file in READ BYTE mode
            file.write(f"{username}:".encode() + encrypted_password + b'\n') #Writing the encoded password in the password.txt file
        print("Password was added successfully in password.txt file")
    except Exception as e:
        print(f"There is error in adding password: {str(e)}")

#*******************Password Retrieve Function**********************************************
def retrieve_password(username, encryption_key):
    try:
        passwords = load_passwords()
        if username in passwords:
            encrypted_password = passwords[username]
            decrypted_password = decrypt_password(encrypted_password, encryption_key)
            return decrypted_password
        else:
            return "Password not found."
    except Exception as e:
        return f"There is error in adding password: {str(e)}"

#*******************Password Updation Function************************************************
def update_password(username, new_password, encryption_key):
    try:
        passwords = load_passwords() #Load password function would load all the passwords in the password.txt file
        if username in passwords: #To check if the username exits 
            encrypted_password = encrypt_password(new_password.encode(), encryption_key)#calling the Encrypt function to encrypt new password
            passwords[username] = encrypted_password
            save_passwords(passwords)
            print("Password was updated successfully.")
        else:
            print("ERR0R: Password was not found.")
    except Exception as e:
        print(f"Error while updating password: {str(e)}")
        
#****************************Function to save passwords*****************************************
def save_passwords(passwords):
    try:
        with open("passwords.txt", "wb") as file: #Open the file using context manager in WRITE BYTE Mode
            for username, encrypted_password in passwords.items(): 
                file.write(f"{username}:".encode() + encrypted_password + b'\n')#Writing the password in the password.txt file
    except Exception as e:
        print(f"Error saving passwords: {str(e)}")

#*****************************Function to load passwords*****************************************
def load_passwords():
    passwords = {}
    try:
        if os.path.exists("passwords.txt"): #To check if the "password.txt" file exists in the path
            with open("passwords.txt", "rb") as file: #Using context manager to open the file in read byte mode
                my_lines = file.readlines() #reading line one by one
                for line in my_lines:
                    seprated = line.strip().split(b":") #Spliting the username and passwords using ":"
                    if len(seprated) == 2:
                        username=seprated[0]
                        encrypted_password =seprated[1]
                        
                        passwords[username.decode()] = encrypted_password
    except Exception as e:
        print(f"There is error loading passwords: {str(e)}")
    return passwords

 
# Main Password Manager logic begins here
def main():
    encryption_key = os.urandom(16)  # To generate random encryption key
    
    while True:
        print("************************* Welcome to Password Manager *********************************")
        print(" ")
        print("                          Please select Options: ")
        print("                          1) Press 1 to Add Password")
        print("                          2) Press 2 to Retrieve Password")
        print("                          3) Press 3 to Update Password")
        print("                          4) Press 4 to Quit!")
        print(" ")
        print("***************************************************************************************")

        
        option = input("Enter your options: ") #Select from four options
        if option == "1":
            username = input("Enter username: ") #Input username from the user
            print("!!!Please make sure that password has maximum 16 characters only!!!")
            password = input("Enter password: ") #Input password from the user
            add_password(username, password, encryption_key)#calling Function to add the password in the password.txt file
        elif option == "2":
            username = input("Enter username: ")
            decrypted_password_bytes = retrieve_password(username, encryption_key)#calling Function to retrieve the password in the password.txt file
            if isinstance(decrypted_password_bytes, bytes):
                decrypted_password = decrypted_password_bytes.decode('utf-8')#decode function to handle character conversion error
                print(f"The Password for the user {username} is : {decrypted_password}")
            else:
                print(decrypted_password_bytes)#Printing the decrypted password
        elif option == "3":
            username = input("Please enter username: ")
            new_password = input("Please enter new password: ") #Enter the password that you want to update
            update_password(username, new_password, encryption_key)#calling function to update the password 
        elif option == "4":
            print("!!!Thanks for using Password Manager!!!")
            break
        else:
            print("Please Enter the choice between (1-4) only!!!")

if __name__ == "__main__":
    main()
```
